From the above information

- There are no missing values
- But ActivityDate is classified as object instead of datetime64

```python
# Convert the ActivityDate to datetime64 data type

daily_activity["ActivityDate"] = pd.to_datetime(daily_activity['ActivityDate'])
```

```python
# Confirm if the dtype changed

daily_activity.info()
```

```python
daily_activity['ActivityDate'].head()
```

Create new list of rearranged columns and renaming the "daily_activity" to df

```python
# Rearranged columns

new_columns = ['Id', 'ActivityDate', 'DayOfTheWeek', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 
              'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalExcerciseMinutes', 'TotalExcerciseHours', 'Calories']
# reindex function to rearrange columns

df = daily_activity.reindex(columns = new_columns)
```

```python
df.head()
```

```python
# Representing the column 'DayOfTheWeek' by day of the week

df['DayOfTheWeek'] = df['ActivityDate'].dt.day_name()

# Confirm the changes

df['DayOfTheWeek'].head()
```

```python
# Renaming the columns
# Using dictionaries

df.rename(columns = {'Id' : 'id', 'ActivityDate' : 'date', 'DayOfTheWeek' : 'day_of_week', 'TotalSteps' : 'total_steps', 
                    'TotalDistance' : 'total_dist', 'TrackerDistance' : 'tracker_dist', 'LoggedActivitiesDistance' : 'logged_dist', 
                    'VeryActiveDistance' : 'very_active_dist', 'ModeratelyActiveDistance' : 'moderately_active_dist', 
                    'LightActiveDistance' : 'light_active_dist', 'SedentaryActiveDistance' : 'sedentary_active_dist', 
                     'VeryActiveMinutes' : 'very_active_mins', 'FairlyActiveMinutes' : 'fairly_active_mins', 'LightlyActiveMinutes' : 'lightly_active_mins', 
                    'SedentaryMinutes' : 'sedentary_mins', 'TotalExcerciseMinutes' : 'total_mins', 'TotalExcerciseHours' : 'total_hours', 'Calories' : 'calories'}, inplace = True)
```

```python
df.head()
```

Updating a column named 'total_mins' to the sum of total time logged

```python
df['total_mins'] = df['very_active_mins'] + df['fairly_active_mins'] + df['lightly_active_mins'] + df['sedentary_mins']

df['total_mins'].head()
```

Convert total_mins to number of hours

```python
# Create a column named 'total_hours' by converting to hour and round float to two decimal places

df['total_hours'] = round(df['total_mins'] / 60)

df['total_hours'].head()
```

```python
# Checking for a number of duplicates

df.duplicated().sum()
```

```python
df.isnull().sum()
```

Data is now ready to be analyzed

```python
# Number of times users logged in across the week

plt.style.use('default')
plt.figure(figsize=(6,4))

plt.hist(df.day_of_week, bins = 7, edgecolor='black')

plt.xlabel('Day of the week')
plt.ylabel('Frequency')
plt.title('No. of times users logged in app across the week')
plt.show()
```

```python
# Sedentary mins vs Total Hours

x = np.array(["sedentary_mins", "lightly_active_mins", "fairly_active_mins", "very_active_mins"])
y = np.array([df["sedentary_mins"].mean(),
             df["lightly_active_mins"].mean(),
             df["fairly_active_mins"].mean(),
             df["very_active_mins"].mean()])

plt.bar(x, y, width = 0.8)
plt.title("Average Minutes")
plt.show()
```

##### . This shows that users spend more time lying down, than they do being active, This can also say something about them spending countless of hours sitting behind their desk working.

```python
# Total calories burnt by day of the week

plt.bar(df.day_of_week, df.calories)
plt.title("Calories Burnt by Day of The Week")
plt.xlabel("Week Day")
plt.ylabel("Calories Burnt")

plt.show()
```

```python
# Total Steps by Day Of The Week

plt.bar(df.day_of_week, df.total_steps)
plt.title("Total Steps In Each Weekday")
plt.xlabel("Week Day")
plt.ylabel("Steps Taken")
plt.show()
```

##### . Most calories are burnt on Saturday and Sunday, which means users are most likely working class individuals.
##### . Most steps are taken on weekends since most users are too busy during the week

```python
# Active Hours Of The Day

fig, axs = plt.subplots(figsize=(10, 6))
df.groupby(df['total_hours'])['total_steps'].mean().plot(kind='bar', rot=0, ax=axs, title='The Most and Least Active Hours of The Day')
plt.xlabel("Hour of the Day")
plt.ylabel("Total Steps")
plt.show()
```

##### . This shows that users are more active in the afternoons
